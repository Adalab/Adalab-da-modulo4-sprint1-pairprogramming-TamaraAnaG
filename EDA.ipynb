```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('HuluRaw.csv', index_col=0)
```

```python
def eda_df(df):

    display(df.info())

    print(f'\n Filas aleatorias: \n')
    display(df.sample(10))

    print("\n Los nulos del conjunto de datos son: \n")
    nulos = pd.DataFrame(df.isnull().sum() / df.shape[0] * 100, columns = ["%_nulos"])
    display(nulos[nulos["%_nulos"] > 0])

    print("\n Los duplicados son: \n")
    display(df.duplicated().sum())

    print(f"Los tipos de las columnas son:")
    display(pd.DataFrame(df.dtypes, columns = ["tipo_dato"]))

    print(f"Lista de los nombres de las variables numricas:")
    display(numericas = df.select_dtypes(include = np.number).columns)

    print("Los valores que tenemos para las columnas categricas son: ")
    categoricas = df.select_dtypes(include = "O")
    for col in categoricas.columns:
        print(f"La columna {col.upper()} tiene las siguientes valores nicos:")
        display(pd.DataFrame(df[col].value_counts()).head())

    print("Los valores que tenemos para las columnas numricas son: ")
    numericas = df.select_dtypes(exclude='O')
    try:
        print("\n ..................... \n")
        print(f"Los principales estadsticos de las columnas categricas son: ")
        display(categoricas.describe(include = "O").T)
    except:
        print('No hay columnas categricas')
    try:
        print("\n ..................... \n")
        print(f"Los principales estadsticos de las columnas numricas son: ")
        display(numericas.describe().T)
    except:
        print('No hay columnas numricas')

    return eda_df

eda_df(df)
```
